# Question 1
* Create a bar chart where x axis displays the values indicating whether customers have multiple lines, and y axis displays the count.
* The goal here is to visualize the count of customers by the nukmber of lines they have.

```python
###
### YOUR CODE HERE
#### Set the style for better visualization
# Set plot style
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='MultipleLines')

# Add labels and title
plt.title('Count of Customers by Multiple Lines')
plt.xlabel('Multiple Lines')
plt.ylabel('Number of Customers')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

# Question 5

# Question 6
* repeat the same analysis. But this time, create seperate histograms of montlycharges for each contract type.
* Recall that there are three contract types month-to-month, One year, and Two year. So you will have three histograms
* You can use plt.subplot and for loops, or any other method.

```python
###
### YOUR CODE GOES HERE
###
# Set up figure with 3 subplots
contract_types = df['Contract'].unique()
plt.figure(figsize=(15, 4))

# Loop through each contract type
for i, contract in enumerate(contract_types, 1):
    plt.subplot(1, 3, i)
    subset = df[df['Contract'] == contract]
    plt.hist(subset['MonthlyCharges'], bins=range(0, int(df['MonthlyCharges'].max()) + 15, 15), edgecolor='black')
    plt.title(f'{contract} Contracts')
    plt.xlabel('Monthly Charges ($)')
    plt.ylabel('Number of Customers')

plt.tight_layout()
plt.show()
```

```python
###
### YOUR INTERPRETATION GOES HERE
###
interpretation = """
These histograms show the distribution of Monthly Charges separately for each Contract Type: Month-to-month, One year, and Two year. Each histogram uses a bin size of $15.

Observations:
- Month-to-month: The distribution is spread widely, with many customers in the lower to mid-range of charges. There’s a long tail indicating some customers have very high charges.
- One year: The charges are more concentrated in the mid-range, with fewer customers paying very high or very low amounts.
- Two year: The distribution is narrowest, with most customers paying between $20 and $70. High-end charges are rare.

Business Insight:
- Month-to-month customers show the most variation in pricing, possibly due to added services or fewer discounts.
- One and Two year customers are more stable in terms of pricing — perhaps due to bundled pricing or contract-based promotions.
- This analysis can help tailor pricing strategies by contract type.
"""

print(interpretation)
```

# Question 10
* Create a heatmap summarizing MonthlyCharges by Sentiment and Contract Type

```python
###
### YOUR CODE GOES HERE
###
# Create a pivot table: average MonthlyCharges for each combination of Sentiment and Contract
heatmap_data = df.pivot_table(values='MonthlyCharges', index='Sentiment', columns='Contract', aggfunc='mean')

# Set figure size
plt.figure(figsize=(10, 5))

# Create the heatmap
sns.heatmap(heatmap_data, annot=True, fmt=".2f", cmap='Blues')

# Add chart title and axis labels
plt.title('Average Monthly Charges by Sentiment and Contract Type')
plt.xlabel('Contract Type')
plt.ylabel('Sentiment')
plt.tight_layout()
plt.show()
```

```python
###
### YOUR INTERPRETATION GOES HERE
###
interpretation = """
This heatmap shows the average Monthly Charges grouped by Sentiment and Contract Type.

Observations:
- Month-to-month customers generally have higher average monthly charges, especially those with Negative or Neutral sentiment.
- Two year contract customers tend to have lower monthly charges, across all sentiment categories.
- There may be a connection between high charges and negative sentiment, particularly in short-term contracts.

Business Insight:
- Customers on month-to-month contracts and expressing negative sentiment may feel they're not getting value for their money.
- Offering tailored pricing, loyalty programs, or better support to these segments could improve satisfaction and reduce churn risk.
"""

print(interpretation)
```
